```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = datasets.iloc[:, [2,3]].values
y  = datasets.iloc[:, 4].values
```

```python
datasets.info()
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
   x, y, test_size= 0.23, random_state=92174
)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
model_RGL = LogisticRegression(random_state=24124)
model_RGL.fit(x_train, y_train)
```

```python
apin = x_test[[10]]
apin
# idin = [[27, 20000]]
```

```python
prediksi = model_RGL.predict(apin)
if prediksi==0:
   print(prediksi, "Berpotensi Tidak Membeli")
else:
   print(prediksi, "Berpotensi Membeli")
prediksi
```
