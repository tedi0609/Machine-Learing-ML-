```python
from bokeh.plotting import figure, output_notebook, show

output_notebook()
p = figure(plot_width = 400, plot_height = 400)
p.circle([1,2,3,4,5],
      [4,7,1,6,3],
      size=10, color="red", alpha=0.5)
show(p)
```

```python
from bokeh.plotting import figure, output_notebook, show

output_notebook()
p = figure(plot_width = 400, plot_height = 400)
p.line([1, 2, 3, 4, 5], [3, 1, 2, 6, 5],
   line_width = 2, color = "green")

show(p)
```

```python
import numpy as np
import matplotlib.pyplot as plt

provinsi = ["NTB", "Jakarta", "Sulsel", "Jabar", "Riau"]
ump = [2200000, 3900000, 2800000, 3200000, 2800000]

plt.bar(provinsi, ump)
plt.title("Perbandingan Besaran Upah Minimum Provinsi")
plt.xlabel("Daftar Nama Provinsi")
plt.ylabel("Upah Minimum")
plt.show()
```
